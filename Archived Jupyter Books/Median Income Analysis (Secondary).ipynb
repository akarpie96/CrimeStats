```python
%matplotlib notebook
import pandas as pd
from scipy.stats import linregress
import matplotlib.pyplot as plt
import numpy as np
```

```python
Median_Inc_df=pd.read_csv("Clean_Data_Output/Clean_Data_Final.csv")
Median_Inc=pd.read_csv("Resources/Median Income.csv")
```

```python
Median_Inc_df["Median Household Income"]=Median_Inc["Median Income (2019)"]
```

```python
Median_Inc_df=Median_Inc_df.sort_values(by="Median Household Income")
```

```python
Median_Inc_df
Median=Median_Inc_df["Median Household Income"].str.replace(",", "").astype(float)


Median_Inc_df["Median Income"]=Median

Median_Inc_df
```

```python
plt.scatter(Median_Inc_df["Median Income"], Median_Inc_df["Murders Per Capita/Homicide Rate"])
(slope, intercept, rvalue, pvalue, stderr) = linregress(Median_Inc_df["Median Income"], Median_Inc_df["Murders Per Capita/Homicide Rate"])
regress_values = Median_Inc_df["Median Income"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.title('Median Household Income relative to Homicide Rate',fontsize =12)
plt.xlabel('Median Household Income',fontsize =10)
plt.ylabel('Homicide Rate',fontsize =10)
plt.plot(Median_Inc_df["Median Income"],regress_values,"r-")
plt.savefig("Image Output/Median_Income_Regression(2).png")
```
