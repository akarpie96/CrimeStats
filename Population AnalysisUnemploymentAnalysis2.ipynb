```python
%matplotlib notebook
import pandas as pd
from scipy.stats import linregress
import matplotlib.pyplot as plt
import numpy as np
```

```python
Pop_df=pd.read_csv('Clean_Data_Output/Population_Data.csv')
new_pop=pd.read_csv('Resources/Census_Pop_2019.csv')
```

```python
Pop_df
```

```python
Pop_df=Pop_df[["State", "Population (2019)", "Total Murders (2019)"]]
```

```python
Bottom_Five_df=Pop_df.sort_values(by="Population (2019)", ascending=True).head()
```

```python
Bottom_Five_df
```

```python
Top_Five_df=Pop_df.sort_values(by="Population (2019)", ascending=False).head()
```

```python
Top_Five_df
```

```python
state=Pop_df["State"]
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(Pop_df["Population (2019)"], Pop_df["Total Murders (2019)"])
regress_values = Pop_df["Population (2019)"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(Pop_df["Population (2019)"], Pop_df["Total Murders (2019)"])
plt.title('State Homicide Count relative to Population',fontsize =12)
plt.xlabel('State Population (millions)',fontsize =10)
plt.ylabel('Homicides (2019)',fontsize =10)
plt.plot(Pop_df["Population (2019)"],regress_values,"r-")
plt.savefig("Image Output/Population_Regression.png")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
```

```python
Pop_df["2019 Population"]=new_pop["2019 Population"]
```

```python
Pop=Pop_df["2019 Population"].str.replace(",", "").astype(float)
```

```python
Pop_df["2019 Population"]=Pop
```

```python
Pop_df
```

```python
Pop_df["Murders Per Capita/Homicide Rate"]=(round(Pop_df["Total Murders (2019)"]/Pop_df["2019 Population"],5))
```

```python
Pop_df=Pop_df[["State", "Total Murders (2019)", "2019 Population", "Murders Per Capita/Homicide Rate"]]
```

```python
Pop_df
```

```python
unemp_df=pd.read_csv('Resources/state_unemployment.csv')
```

```python
Pop_df["Unemployment Rate"]=unemp_df["unemployment_rate"]
```

```python
Pop_df
```

```python
Pop_df=Pop_df.sort_values(by="Unemployment Rate")
```

```python
plt.scatter(Pop_df["Unemployment Rate"], Pop_df["Murders Per Capita/Homicide Rate"])
(slope, intercept, rvalue, pvalue, stderr) = linregress(Pop_df["Unemployment Rate"], Pop_df["Murders Per Capita/Homicide Rate"])
regress_values = Pop_df["Unemployment Rate"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.title('State Homicide Rate relative to Unemployment Rate',fontsize =12)
plt.xlabel('Unemployment Rate',fontsize =10)
plt.ylabel('Homicide Rate',fontsize =10)
plt.plot(Pop_df["Unemployment Rate"],regress_values,"r-")
plt.savefig("Image Output/Population_Regression(2).png")
```
