```python
#load dependencies
import pandas as pd
from matplotlib.pyplot as plt
import numpy as np
```

```python
#read state population csv ('Population_Data.csv')
Pop_df = pd.read_csv('Clean_Data_Output/Population_Data.csv')
Pop_df.head()
```

```python
#remove column 'Unnamed:0'
Pop_df = Pop_df[["State", "Population (2019)", "Total Murders (2019)"]]
```

```python
#set variable 'Bottom_Five_df' to 5 lowest state populations
Bottom_Five_df=Pop_df.sort_values(by="Population (2019)", ascending=True).head()
Bottom_Five_df
```

```python
#set variable 'Top_Five_df' to 5 highest state populations
Top_Five_df=Pop_df.sort_values(by="Population (2019)", ascending=False).head()
Top_Five_df
```

```python
state=Pop_df["State"]
```

```python
#plot state populations in relaiton to total state homicides as scatterplot
plt.scatter(Pop_df["Population (2019)"], Pop_df["Total Murders (2019)"])
plt.title('State Homicide Count relative to Population',fontsize =12)
plt.xlabel('State Population (millions)',fontsize =10)
plt.ylabel('Homicides (2019)',fontsize =10)
```

```python
#plot 5 lowest state populations as histogram
U_rate = Bottom_Five_df["Population (2019)"]
x_axis = Bottom_Five_df["State"]
plt.bar(x_axis, U_rate, color='b', alpha=0.70, align="center")
plt.xticks(rotation = 45)
plt.ylabel('State Population')
plt.show()
```

```python
#plot 5 highest state populations as histogram
U_rate = Top_Five_df["Population (2019)"]
x_axis = Top_Five_df["State"]
plt.bar(x_axis, U_rate, color='r', alpha=0.70, align="center")
plt.xticks(rotation = 45)
plt.ylabel('State Population')
plt.show()
```
