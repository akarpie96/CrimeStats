```python
import pandas as pd
import numpy as np
import requests
import json
from pprint import pprint
# Google API Key
from config import gkey
import gmaps
gmaps.configure(api_key=gkey)
!pip install gmaps
```

```python
New_df=pd.read_csv('Clean_Data_Output/Data.csv')
```

```python
del New_df["Unnamed: 0"]
```

```python
State_df=New_df["State"]
```

```python
State_df=pd.DataFrame(State_df)
```

```python
State_df["Lat"]=""
State_df["Lng"]=""
```

```python
State_df
```

```python
params = {"key": gkey}

for index, row in State_df.iterrows():
# Loop through the cities_pd and run a lat/long search for each city
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    state = row['State']
    params['address'] = f"{state}"


# make request
    cities_lat_lng = requests.get(base_url, params=params)

    cities_lat_lng = cities_lat_lng.json()

    State_df.loc[index, "Lat"] = cities_lat_lng["results"][0]["geometry"]["location"]["lat"]
    State_df.loc[index, "Lng"] = cities_lat_lng["results"][0]["geometry"]["location"]["lng"]
#Verify Result
State_df.head()
```

```python
State_df["Total Murders (2019)"]=New_df["Total Murders (2019)"]

```

```python
State_df
```

```python
Heat_map_df=State_df
```

```python
Heat_map_df
```

```python
fig = gmaps.figure()
locations=Heat_map_df[["Lat", "Lng"]]
murders=Heat_map_df["Total Murders (2019)"]
# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=murders, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=2)


# Add layer
fig.add_layer(heat_layer)

# Display figure

```
