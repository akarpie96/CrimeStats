```python
#load dependencies
import pandas as pd
```

```python
#Conducting analysis of relationship between State Population and Homicide Count for 2019

#Reading in csv files for population estimate and homicide figures. 
population_df=pd.read_csv('Resources/Census_Pop_2019.csv')
homicide_df=pd.read_csv('Resources/FBI_Homicide_Data.csv')
median_df=pd.read_csv('Resources/Median Income.csv')
```

```python
median_df.head()
```

```python
population_df.head()
```

```python
#Cleaning Homicide_df

homicide_df.dropna(how='all')
#Adding column from homicide data frame to population Data Frame 
homicide_df.head()
```

```python
#Removing unneccesarry columns

homicide_df=homicide_df.head(51)

homicide_df=homicide_df.rename(columns={"Total\nmurders1": "Total Murders", "Total\nfirearms":"Total Firearms", "Firearms\n(type\nunknown)":"Firearms (type uknown)", "Knives or\ncutting\ninstruments":"Knives or cutting Instruments", "Other\nweapons":"Other Weapons", "Hands, fists,\nfeet, etc.2": "Hands, fists, feet, etc"})

homicide_df
```

```python
#dsiplay cleaned homicide data frame ('homicide_df')
homicide_df.head()
```

```python
#remove row 'United States' from median income data frame ('median_df')
median_df=median_df.loc[median_df["State"]!="United States", :]

median_df["2019 Population"]=population_df["2019 Population"]

median_df=median_df.reset_index()

median_df=median_df[["State", "Median Income (2019)", "2019 Population"]]

median_df.head()
```

```python
population=median_df['2019 Population'].str.replace(",", "").astype(float)
population.head()
```

```python
median_df["Population (2019)"]=population

total_murders=homicide_df["Total Murders"].str.replace(",", "").astype(float)

median_df["Total Murders (2019)"]=total_murders

median_df.head()
```

```python
Population_df=median_df[["State", "Population (2019)", "Total Murders (2019)"]]


Summary_df=median_df


Summary_df=Summary_df[["State", "Median Income (2019)", "2019 Population", "Population (2019)", "Total Murders (2019)"]]

Summary_df.to_csv("Clean_Data_Output/Data.csv", index=True)

Population_df.to_csv("Clean_Data_Output/Population_Data.csv", index=True)
```
